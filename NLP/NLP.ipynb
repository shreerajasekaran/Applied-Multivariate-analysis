```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

# 1. Create a tiny toy dataset of "Student AI Tool Reviews"
reviews = [
    "Gemini is incredibly fast and helpful for my coding.",
    "I hate how Copilot suggests wrong syntax, it is terrible.",
    "Claude writes beautifully and saves me so much time!",
    "Perplexity crashed and deleted my research, very bad experience.",
    "ChatGPT is a lifesaver for my essays, absolutely love it.",
    "The AI tool is useless and completely hallucinated the facts.",
    "Such an amazing and smart assistant, highly recommend.",
    "Worst tool ever, totally ruined my assignment."
]
# Labels: 1 = Positive, 0 = Negative
labels = [1, 0, 1, 0, 1, 0, 1, 0]

df_nlp = pd.DataFrame({'Review': reviews, 'Sentiment': labels})

# 2. Translate Text to Numbers (Bag of Words)
# Computers can't read text, so we count how many times each word appears
vectorizer = CountVectorizer(stop_words='english')
X_vectors = vectorizer.fit_transform(df_nlp['Review'])

# Let's peek at the "Vocabulary" it learned
vocab = vectorizer.get_feature_names_out()

# 3. Train a quick Classification Model (Naive Bayes)
model = MultinomialNB()
model.fit(X_vectors, df_nlp['Sentiment'])

# 4. Test the model on completely unseen sentences!
test_sentences = [
    "This AI is amazing and fast!",
    "I had a terrible and bad experience with this useless tool."
]

X_test = vectorizer.transform(test_sentences)
predictions = model.predict(X_test)

print("--- NLP SENTIMENT ANALYSIS PIPELINE ---")
print(f"Total words learned in vocabulary: {len(vocab)}\n")

print("PREDICTIONS ON NEW TEXT:")
for sentence, pred in zip(test_sentences, predictions):
    sentiment = "POSITIVE 😃" if pred == 1 else "NEGATIVE 😡"
    print(f"Text: '{sentence}'")
    print(f"Predicted Sentiment: {sentiment}\n")
```

```python
import pandas as pd
import nltk
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer

# Quietly download necessary NLTK dictionary files
nltk.download('wordnet', quiet=True)
nltk.download('omw-1.4', quiet=True)

# Let's pick a list of tricky words
words = ["studies", "caring", "foxes", "mice", "better", "ran"]

# Initialize our two tools
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

# We will specify the Part of Speech (POS) for verbs/adjectives so the lemmatizer works perfectly
pos_tags = ['n', 'v', 'n', 'n', 'a', 'v'] # n=noun, v=verb, a=adjective

results = []
for i in range(len(words)):
    original = words[i]
    stemmed = stemmer.stem(original)
    lemmatized = lemmatizer.lemmatize(original, pos=pos_tags[i])
    results.append([original, stemmed, lemmatized])

df_nlp_compare = pd.DataFrame(results, columns=["Original Word", "Stemmed (Machete)", "Lemmatized (Dictionary)"])
print(df_nlp_compare.to_string(index=False))
```
