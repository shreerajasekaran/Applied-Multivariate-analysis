```python
import pandas as pd
import scipy.stats as st
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df = pd.read_csv("random_50_samples.csv")
```

```python
fig = px.box(
    df,
    x="Admission Type",
    y="Billing Amount",
    color="Admission Type",
    title="Billing Amount by Admission Type",
    labels={"Billing Amount": "Bill Amount", "Admission Type": "Admission Type"},
)
fig.show()
```

```python
# Our Null hypothesis is that there is no significant difference in billing amounts between the different admission types. The alternative hypothesis is that there is a significant difference in billing amounts between at least two of the admission types.

alpha = 0.05

elective_billing = df[df["Admission Type"] == "Elective"]["Billing Amount"]
emergency_billing = df[df["Admission Type"] == "Emergency"]["Billing Amount"]
urgent_billing = df[df["Admission Type"] == "Urgent"]["Billing Amount"]

anova, p_value = st.f_oneway(elective_billing, emergency_billing, urgent_billing)

print(f"ANOVA Test Statistic: {anova:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Alpha: {alpha}")
print()
if p_value < alpha:
    print("Result: Reject the Null Hypothesis.")
    print(
        "There IS a significant difference in billing amounts between the different admission types."
    )
else:
    print("Result: Fail to Reject the Null Hypothesis.")
    print(
        "There is NO significant difference in billing amounts between the different admission types."
    )
```

```python
# Let's check the assumptions of ANOVA before we conclude. We need to check for normality and homogeneity of variances.
# Levene's test for equal variance
levene_stat, levene_p = st.levene(elective_billing, urgent_billing, emergency_billing)
print(f"Levene's Test p-value: {levene_p:.4f}")


# Shapiro-Wilk test for normality
print(st.shapiro(elective_billing).pvalue)
print(st.shapiro(urgent_billing).pvalue)
print(st.shapiro(emergency_billing).pvalue)
```
