```python
import pandas as pd
import scipy.stats as st
import matplotlib.pyplot as plt
import plotly.express as px
import numpy as np
```

```python
df = pd.read_csv("random_50_samples.csv")
```

```python
# Let's plot and explore!
contingency = pd.crosstab(df["Gender"], df["Admission Type"])

labels = contingency.index
elective = contingency["Elective"]
emergency = contingency["Emergency"]
urgent = contingency["Urgent"]

x = np.arange(len(labels))  # the label locations
width = 0.25  # the width of the bars

# Create the grouped bar chart using only matplotlib
fig, ax = plt.subplots(figsize=(8, 6))
rects1 = ax.bar(x - width, elective, width, label="Elective")
rects2 = ax.bar(x, emergency, width, label="Emergency")
rects3 = ax.bar(x + width, urgent, width, label="Urgent")

ax.set_ylabel("Number of Patients")
ax.set_xlabel("Gender")
ax.set_title("Admission Type Breakdown by Gender")
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend(title="Admission Type")
```

```python
alpha = 0.05
chi2_stat, p_val, dof, expected = st.chi2_contingency(contingency)
print(f"Chi-square statistic: {chi2_stat:.4f}")
print(f"p-value: {p_val:.4f}")

if p_val < alpha:
    print(
        "Reject the null hypothesis: There is a significant association between gender and admission type."
    )
else:
    print(
        "Fail to reject the null hypothesis: There is no significant association between gender and admission type."
    )
```
