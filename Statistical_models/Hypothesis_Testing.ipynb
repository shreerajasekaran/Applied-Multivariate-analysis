```python
# In this notebook, we will perform hypothesis testing to determine if there is a significant difference in billing amounts based on Gender. I have taken this data set from Kaggle and it had a large sample size of more than 1000 records. I randomly selected 50 records from the data set to perform the hypothesis testing.
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
import scipy.stats as st
```

```python
df = pd.read_csv("random_50_samples.csv")
df
```

```python
fig = px.box(
    df,
    x="Gender",
    y="Billing Amount",
    color="Gender",
    title="Billing Amount by Gender",
    labels={"Billing Amount": "Bill Amount", "Gender": "Gender"},
)
fig.show()
```

```python
df["Billing Amount"].describe()
```

```python
# Hypothesis Testing:
# Null Hypothesis (H0): There is no significant difference in billing amounts based on Gender.
# Alternative Hypothesis (H1): There is a significant difference in billing amounts based on Gender
```

```python
df.groupby("Gender")["Billing Amount"].describe()
```

```python
# In this case I have used Welch's t-test because it does not assume equal variances between the two. There was not a huge difference in the standard deviations of the two groups, but I wanted to be cautious given the small sample size.

alpha = 0.05

male_billing = df[df["Gender"] == "Male"]["Billing Amount"]
female_billing = df[df["Gender"] == "Female"]["Billing Amount"]

t_stat, p_value = st.ttest_ind(male_billing, female_billing, equal_var=False)

print(f"Welch's t-test Results")
print(f"----------------------")
print(f"T-statistic : {t_stat:.4f}")
print(f"P-value     : {p_value:.4f}")
print(f"Alpha       : {alpha}")
print()
if p_value < alpha:
    print("Result: Reject the Null Hypothesis.")
    print(
        "There IS a significant difference in billing amounts between Male and Female."
    )
else:
    print("Result: Fail to Reject the Null Hypothesis.")
    print(
        "There is NO significant difference in billing amounts between Male and Female."
    )
```
